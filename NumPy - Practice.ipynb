ðŸ”¥ Mini Project: Temperature Data Analysis

ðŸ§  Real-World Scenario

A weather station recorded temperatures for 7 days:

30, 32, 35, 28, 40, 38, 31

You need to analyze:

-->Average temperature

-->Days above average

-->Hottest and coldest day

-->Temperature range

-->Normalize temperatures (scale between 0 and 1)

This is similar to real data preprocessing before ML.

```python
import numpy as np

temps = np.array([30, 32, 35, 28, 40, 38, 31])

# Find average temperature.
# Find which days are above average.
# Find:
# --Maximum temperature
# --Minimum temperature
# Find temperature range.
# Normalize temperatures between 0 and 1 using:

avg_temp = np.mean(temps)
days_above_avg = np.where(temps > avg_temp)
max_temp = np.max(temps)
min_temp = np.min(temps)
temp_range = max_temp - min_temp
nomarlize_temp = (temps - min_temp) /temp_range



print("Average temperature:-",avg_temp)
print("The days are above average:-",days_above_avg[0]+1)
print("Maximum temperature:-",max_temp)
print("Minimum temperature:-",min_temp)
print("Temperature range:-", temp_range)
print("Normalize temperatures:-",nomarlize_temp)


```

# TASK - 3 (REAL WORLD PROBLEM SOLVING) DATASETS

```python
# Dataset: 5 employees, each with monthly salary
# salaries = [25000, 55000, 72000, 38000, 90000]

# 1. Using np.where() â€” label each as "Junior" if below 50000, else "Senior"
# 2. Create a 2D array of shape (5, 2) where column 1 = salaries, 
#    column 2 = salaries after 20% hike (salary * 1.2)
#    Hint: use reshape and hstack
# 3. From that 2D array, extract only the "after hike" column (column index 1)
# 4. Find the max, min, and mean of the after-hike salaries
# 5. Reshape the original salaries array into shape (5, 1)


salaries = np.array([25000, 55000, 72000, 38000, 90000])

find_jun_sen = np.where(salaries < 50000 ,"Junior" , "Senior" )

hike_of_20 = salaries * 1.2
sal_hike20 = np.hstack((salaries.reshape(5,1) , hike_of_20.reshape(5,1)))  

# extracting only after hike column
extract_hike_col = sal_hike20[:,1]

# max, min, and mean of the after-hike salaries
max_hike = np.max(extract_hike_col)
min_hike = np.min(extract_hike_col)
mean_hike = np.mean(extract_hike_col)

# Reshape the original salaries array into shape (5, 1)
reshape_salaries = salaries.reshape(5,1)

print("Finding Juniors and Seniors in Employees:-", find_jun_sen)
print("Create a 2D array of shape (5, 2) where column 1 = salaries and column 2 = hike salaries:-\n",sal_hike20)
print("Extracting the only Hike column in the 2D array:-",extract_hike_col)
print("Maximum hike salary:-",max_hike)
print("Minimum hike salary:-",min_hike)
print("Average hike salary:-",mean_hike)
print("Reshape the original salaries array:-\n",reshape_salaries)
```

```python
# A class has 6 students with their marks in two subjects:
# Math:    [55, 80, 45, 90, 60, 72]
# Science: [70, 65, 80, 85, 50, 78]

# 1. Find the total marks of each student (math + science)
# 2. Find the average marks of each student
# 3. Label each student as "Distinction" if average >= 75, else "Pass" if >= 50, else "Fail"
#    Hint: you can use nested np.where() for 3 conditions
# 4. Find how many students got Distinction
# 5. Find the name index of the student with highest total marks

import numpy as np

math = np.array([55, 80, 45, 90, 60, 72])
science = np.array([70, 65, 80, 85, 50, 78])


math_science_2d = np.vstack((math , science)).reshape(2,6)
total_marks_stud = np.sum(math_science_2d , axis = 0)           # 1. Find the total marks of each student (math + science)
average_stud = np.mean(math_science_2d , axis = 0)              # 2. Find the average marks of each student


# 3. Label each student as "Distinction" if average >= 75, else "Pass" if >= 50, else "Fail"
finding_grade = np.where(average_stud >= 75, "Distinction", np.where(average_stud >=50 ,"Pass" , "Fail"))        


# finding_students = np.where()

stud_got_dist = np.sum(average_stud >= 75)     # 4. Find how many students got Distinctionâ­â­â­â­

stud_got_dist_way_2 = len(np.where(average_stud >=75)[0])   #Another way to find the count â­â­â­â­

index_of_stud = np.argmax(total_marks_stud) + 1     # 5. Find the name index of the student with highest total marks


print("Total marks of each student (math + science):-",total_marks_stud)
print("Average marks of each student:-",average_stud)
print("Finding the Grade of the student :-",finding_grade)
print("Finding students got Distinction :-",stud_got_dist_way_2)
print("Index of the student with highest total marks :-",index_of_stud)
```

```python
# A shop recorded daily sales for 2 weeks (14 days):

# Week 1: [1200, 1500, 1100, 1800, 2000, 2500, 3000]
# Week 2: [1400, 1600, 1000, 1900, 2100, 2800, 3200]

# 1. Stack Week 1 and Week 2 into a 2D array of shape (2, 7)
# 2. Find total sales for each week (sum of each row)
# 3. Find which day of the week had highest average sales across both weeks
# 4. Find all days where sales were above 2000 (use np.where to get indices)
# 5. Add a 10% bonus to all days where sales were below 1500
#    Hint: use np.where() 3-argument form
# 6. Reshape the entire 14-day data into shape (7, 2) â€” each row = one day, columns = week1 and week2
import numpy as np

week_1 = np.array([1200, 1500, 1100, 1800, 2000, 2500, 3000])
week_2 = np.array([1400, 1600, 1000, 1900, 2100, 2800, 3200])

two_weeks = np.hstack([week_1 , week_2]).reshape(2,7)

# Find total sales for each week (sum of each row)
total_week_sales = np.sum(two_weeks , axis = 1)


# Find which day of the week had highest average sales across both weeks
average_sales = np.mean(two_weeks , axis = 0)
which_day = np.argmax(average_sales)+1

# Find all days where sales were above 2000 (use np.where to get indices)
week_1_week_2 = np.hstack((week_1 ,week_2))
sales_days_indices = np.where(week_1_week_2 >2000)
values = week_1_week_2[sales_days_indices]


# 5. Add a 10% bonus to all days where sales were below 1500
#    Hint: use np.where() 3-argument form
hike_sales_days = np.where(week_1_week_2 <1500 , week_1_week_2 * 1.10 , week_1_week_2)



# 6. Reshape the entire 14-day data into shape (7, 2) â€” each row = one day, columns = week1 and week2
reshaping_14days = np.hstack([week_1.reshape(7,1) , week_2.reshape(7,1)])




print("Stack Week 1 and Week 2 into a 2D array of shape (2, 7):-\n",two_weeks)
print("Total sales for each week:-",total_week_sales)
print("day of the week had highest average sales across both weeks:-",which_day)
print("all days where sales were above 2000:-",values)
print("10% bonus to all days where sales were below 1500:-",hike_sales_days)
print("Reshape the entire 14-day data into shape (7, 2):-\n",reshaping_14days)


```
