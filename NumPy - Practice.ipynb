```python
import numpy as np

marks = np.array([
    [78, 85, 90],
    [88, 79, 92],
    [65, 70, 72],
    [90, 92, 94]
])

# Find total marks of each student
total_marks = np.sum(marks , axis = 1)
print("Total marks of each student :-",total_marks)

# Find lowest marks in each subject
lowest_marks = np.min(marks , axis = 0)
print("Lowest marks in each subject :-",lowest_marks)


# Find overall class average
average = np.mean(marks)
print(average)
```

Tasks for You

Find average salary of each employee

Find average salary of each month

Give every employee a 10% increment using broadcasting

Find the new average salary of each employee after increment

```python

salary = np.array([
    [25000, 26000, 27000],
    [30000, 31000, 32000],
    [20000, 21000, 22000],
    [40000, 42000, 43000]
])

average_sal_emp = np.mean(salary , axis = 1)
print("Average salary of each employee :-", average_sal_emp)

average_sal_mon = np.mean(salary , axis = 0)
print("Average salary of each month", average_sal_mon)

increment = salary * 1.10
print("The increment of 10% of every employee :-", increment)

average_sal_inc = np.mean(increment , axis =1)
print("New average salary of each employee after increment", average_sal_inc)

highest_sal = np.max(increment , axis = 0)
print("Highest salary of each month :-",highest_sal)

high_total_sal = np.sum(increment , axis= 1)
print("Employ with highest total salary :-" , high_total_sal)
```

```python
# Find the highest sale in each day
# Find the lowest sale in each product
# Find overall average sales

import numpy as np

sales = np.array([
    [120,150,100,130],
    [140,160,110,120],
    [130,170,120,140],
    [150,180,130,150],
    [160,190,140,160]
])


high_sale = np.max(sales , axis = 1)
print("Highest sale in each day:-",high_sale)

low_sale = np.min(sales , axis = 0)
print("lowest sale in each product",low_sale)

average_sales = np.mean(sales)
print("overall average sales:-",average_sales)
```

```python
stock = np.array([
    [50, 40, 30, 20],
    [60, 45, 35, 25],
    [55, 50, 40, 30],
    [65, 55, 45, 35],
    [70, 60, 50, 40]
])


# Total stock per product
# Total stock per week
# Product with highest stock
# Week with lowest stock
# Find the highest stock recorded anywhere in the table.
total_pro = np.sum(stock , axis = 0)
total_stock = np.sum(stock , axis = 1)

product_with_highest = np.argmax(total_pro)
week_with_lowest = np.argmin(total_stock)

highest_stock = np.max(stock)
# position = np.unravel_index(np.argmax(stock), stock.shape)


print("Total stock per product:", total_pro)
print("Total stock per week:", total_stock)
print("Product with highest total stock: Product", product_with_highest + 1)
print("Week with lowest total stock: Week", week_with_lowest + 1)
print("Highest stock recorded:", highest_stock)
# print("Position of highest stock (row, col):", position)
```
