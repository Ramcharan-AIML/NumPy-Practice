```python
# Create an array and do the properties of an array

arr_3 = np.array([1,2,3,4,5])

np.shape(arr_3)
np.size(arr_3)
np.ndim(arr_3)

# print(arr_3)
```

```python
# Task 1

# Create numbers 1–16 and reshape into 4×4.

arr_2D = np.arange(1,17).reshape(4,4)

print(arr_2D)
```

```python
# Task 2

# Create numbers 1–9 and reshape into 3×3.

arr_2D_2 = np.arange(1,10).reshape(3,3)

print(arr_2D_2)
```

```python
# Task 3

# Flatten a 3×3 array.

flatten_arr = arr_2D.flatten()

print(flatten_arr)
```

```python
# Task 4

# Create a 2×3 array and print sum of all elements.

sum_arr = np.arange(1 , 7).reshape(2,3)

print(sum_arr)
print(np.sum(sum_arr))
```

```python
# Task 5

# Find row-wise sum using axis=1.

arr_3d = np.arange(1 , 21).reshape(4,5)

row_sum = np.sum(arr_3d , axis = 1)

print(arr_3d)
print(row_sum)
```

```python
# Task 6

# Find column-wise sum using axis=0.

arr_3d2 = np.arange(10 , 40).reshape(5,6)

column_sum = np.sum(arr_3d2 , axis = 0)

print(arr_3d2)
print(column_sum)
```

```python
# Task 7

# Create a marks table (3 students, 4 subjects) and find:
# average of each student

marks_table = np.arange(30 , 42).reshape(3, 4)

average = np.mean(marks_table , axis = 1 )

print(marks_table)
print("average of each student-",average)
```

```python
# Task 8

# Using same table, find:
# highest marks in each subject

highest = np.max(marks_table, axis = 0)
print("Highest marks in each subject - ",highest)
```

```python
# Task 9

# Create a random 3×3 array and reshape into 1×9

random_arr = np.random.randint(10 , 19 , size= (3,3))

reshape_arr = random_arr.reshape(1,9)

print(random_arr)
print(reshape_arr)
```

```python
# Task 10 (Real-world)

# Create salary data of 5 employees for 3 months:
# Find:
# average salary per employee
# average salary per month


salary_data = np.arange(1,16).reshape(5 , 3)

avg_sal = np.mean(salary_data , axis = 1 , dtype = int)
avg_mon = np.mean(salary_data , axis = 0 , dtype = int)

print(salary_data)
print("Average Salary -", avg_sal)
print("Average Month -", avg_mon)
```
